```python
#Python Pandas Data Structure
#DataFrame
import pandas as pd  
# a list of strings  
x = ['Python', 'Pandas']  
# Calling DataFrame constructor on list  
df = pd.DataFrame(x)  
print(df)
```

```python
#Pandas Slicing
#Pandas loc vs. iloc 
#if I wants a perticular row and column from the dataset's
from pandas import DataFrame
df = DataFrame(data)
df1=df[['P1']]
print(df1)
df2=df[['P1','P2','P3']]
print(df2)
```

```python
## retrieving rows by iloc method 
row2 = data.iloc[3] 
print(row2)
```

```python
#Retrieving all rows and some columns by loc method
df3= data.loc[:, ["P1","P2","P3"]]
print(df3)
```

```python
## retrieving all rows and some columns by iloc method 
row3 = data.iloc [:, [0, 1, 2]]
print(row3)
```

```python
# retrieving two rows and two columns by iloc method 
row3 = data.iloc [[3, 4], 
                  [1, 2]]
print(row3)
```

```python
# retrieving some rows and all the columns
row4 = data.iloc [1:5,:]
print(row4)
```
