```python
import numpy as np
import matplotlib.pyplot as plt
```

**Plotting 3-D lines & plots.**

Plotting 3-D lines and plots involves using the plot() function with
three-dimensional data points. This allows for the creation of lines and
curves in three-dimensional space, which can be useful for visualizing
relationships between three variables.

```python
#create data for plot
z=np.linspace(0,np.pi*2)
x=np.sin(z)
y=np.cos(z)
ax=plt.figure().add_subplot(projection='3d')
ax.plot(x,y,z)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.show()
```

**Scatter Plot**

Scatter plots can also be created in three-dimensional space using the
scatter() function. This allows for the visualization of individual data
points and can be useful for identifying patterns or clusters in the data.

```python
# create data for the scatter plot
x = np.random.normal(size=100)
y = np.random.normal(size=100)
z = np.random.normal(size=100)
colors = np.random.rand(100)
# create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# plot the scatter plot
ax.scatter(x, y, z, c=colors, cmap='viridis')
# set the labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

**Surface plot**

Surface plots involve the creation of a three-dimensional surface from
a set of data points. This can be useful for visualizing complex
functions or for identifying areas of high or low density in the data.

```python
# create data for the surface plot
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(np.sin(X)**2 + np.cos(Y)**2)
Z = np.tan(R)
# create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# plot the surface plot
ax.plot_surface(X, Y, Z)
# set the labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
# show the plot
plt.show()
```

**Density & Contour plots.**

Density and contour plots are two techniques that can be used to
visualize the distribution of data in three-dimensional space. Density
plots use color to represent the density of data points in different areas
of the plot, while contour plots use contour lines to represent areas of
equal density.

```python
# create data for the density and contour plot
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
z = (x**2+y**2)/np.pi
# create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# plot the density and contour plot
hist, xedges, yedges = np.histogram2d(x, y, bins=30, density=True)
X, Y = np.meshgrid(xedges[:-1], yedges[:-1])
Z = hist.T
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.contour(X, Y, Z, offset=-0.15, cmap='viridis')
# set the labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
#show plot
plt.show()
```

**Changing the camera angle of 3-D plot.**

To change the camera angle of a 3-D plot in Matplotlib, the view_init()
function can be used. This function allows the user to set the azimuthal
and elevation angles of the plot, controlling the viewpoint from which
the plot is viewed.

```python
#create data for plot
z=np.linspace(0,2*np.pi)
x=np.sin(z)
y=np.cos(z)
ax=plt.figure().add_subplot(projection='3d')
ax.plot(x,y,z)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
# change the camera angle
ax.view_init(elev=45,azim=45)
# show the plot
plt.show()
```

**Creating multiple subplots**

Creating multiple subplots in Matplotlib can be done using the
subplots() function. The subplots() function can be used to create a
grid of subplots, with each subplot containing a different plot or
visualization.

```python
# create some data for the subplots
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
# create a figure with 3 subplots in a vertical layout
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(6, 8))
# plot each subplot
ax[0].plot(x, y1, color='blue')
ax[0].set_title('Sine Function')
ax[1].plot(x, y2, color='green')
ax[1].set_title('Cosine Function')
ax[2].plot(x, y3, color='red')
ax[2].set_title('Tangent Function')
# set the labels for the x and y axes of the figure
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.02, 0.5, 'Y-axis', va='center', rotation='vertical')
# adjust the layout of the subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3)
# show the plot
plt.show()
```

**Stacking subplots in one direction.**

Stacking subplots in one direction involves using the subplot2grid()
function to create a grid of subplots in a single direction, either
vertically or horizontally.

```python
# create a figure with a 3x1 grid of subplots
fig = plt.figure(figsize=(6, 8))
ax1 = plt.subplot2grid((3, 1), (0, 0))
ax2 = plt.subplot2grid((3, 1), (1, 0))
ax3 = plt.subplot2grid((3, 1), (2, 0))
# plot each subplot
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Function')
ax2.plot(x, y2, color='green')
ax2.set_title('Cosine Function')
ax3.plot(x, y3, color='red')
ax3.set_title('Tangent Function')
# set the labels for the x and y axes of the figure
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.02, 0.5, 'Y-axis', va='center', rotation='vertical')
# adjust the layout of the subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3)
# show the plot
plt.show()
```

**Vertically & Horizontally stacked sub plots.**

Vertically and horizontally stacked subplots can be created using the
subplots() function, which allows for the creation of a grid of subplots
in either direction.

```python
# Create a figure with two vertically stacked subplots
fig, axs = plt.subplots(1, 2, figsize=(6, 6))
# Plot the first subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sine Function')
# Plot the second subplot
axs[1].plot(x, y2, color='green')
axs[1].set_title('Cosine Function')
# Set the labels for the x and y axes of the figure
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.01, 0.5, 'Y-axis', va='center', rotation='vertical')
# Adjust the layout of the subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3)
# Show the plot
plt.show()
```

**Stacking subplots in two directions.**

Stacking subplots in two directions involves using the subplot2grid()
function to create a grid of subplots in two directions, with each
subplot containing a different plot or visualization.

```python
# Create a figure with subplots stacked in two directions
fig = plt.figure(figsize=(10, 8))
# Create the first subplot
ax1 = plt.subplot2grid((3, 2), (0, 0))
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Function')
# Create the second subplot
ax2 = plt.subplot2grid((3, 2), (0, 1))
ax2.plot(x, y2, color='green')
ax2.set_title('Cosine Function')
# Create the third subplot
ax3 = plt.subplot2grid((3, 2), (1, 0), colspan=2)
ax3.plot(x, y3, color='red')
ax3.set_title('Tangent Function')
# Create the fourth subplot
ax4 = plt.subplot2grid((3, 2), (2, 0), colspan=2)
ax4.plot(x, y1, color='blue')
ax4.plot(x, y2, color='green')
ax4.plot(x, y3, color='red')
ax4.set_title('All Functions')
# Set the labels for the x and y axes of the figure
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.02, 0.5, 'Y-axis', va='center', rotation='vertical')
# Adjust the layout of the subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3)
# Show the plot
plt.show()
```

**Sharing of the Axes.**

Sharing of the axes can be done using the sharex and sharey
arguments in the subplots() function. This allows for the sharing of the
x- or y-axis between subplots, making it easier to compare different
plots or visualizations.

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1,sharex=True)
# Plot the data on the subplots
ax1.plot(x, y1, color='blue')
ax1.set_ylabel('Sin(x)')
ax2.plot(x, y2, color='green')
ax2.set_ylabel('Cos(x)')
# Add a title to the figure
fig.suptitle('Sine and Cosine Functions')
# Show the plot
plt.show()
```
