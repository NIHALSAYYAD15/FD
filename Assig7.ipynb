## Bivariate analysis

>Bivariate plots are graphs or charts that show the relationship between two variables. They are used to visualize the association between two numerical variables or one numerical variable and one categorical variable. There are several types of bivariate plots, including scatter plots, line graphs, and bar graphs. Scatter plots are one of the most commonly used bivariate plots, and they are used to show the relationship between two continuous variables. In a scatter plot, each data point is plotted as a point on the graph, with one variable represented on the x-axis and the other variable represented on the y-axis. Line graphs are similar to scatter plots, but they are used to show the relationship between two variables over time. In a line graph, the data points are connected by a line to show the trend over time. Bar graphs are used to show the relationship between one numerical variable and one categorical variable. In a bar graph, the numerical variable is represented on the y-axis and the categories are represented on the x-axis. Bivariate plots are an important tool for data analysis and can provide valuable insights into the relationship between variables.

```python
# Bivariate analysis
# Scatter plot of two variables
sns.scatterplot(data=df, x='Age', y='Income')
plt.title('Scatter plot of Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

## DBSACN Algorithm



>DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm used to group together data points that are close to each other in a high-dimensional space. The algorithm works by grouping together points that are within a specified distance (epsilon) of each other, and that have a minimum number of neighbors within that distance. The basic steps of the DBSCAN algorithm are:

1. Choose a point at random and find all the points that are within a distance epsilon of it.
2. If the number of points within epsilon is less than a specified threshold, mark the point as noise and move to the next point. Otherwise, create a new cluster and add all the points within epsilon to it.
3. For each point within the cluster, repeat the process of finding all the points within epsilon and adding them to the cluster. Continue until no more points can be added.
4. Move to the next unvisited point and repeat the process.

>DBSCAN has some advantages over other clustering algorithms, such as its ability to handle noise and outliers, and its ability to identify clusters of arbitrary shape. However, it can be sensitive to the choice of parameters (epsilon and the minimum number of points) and can be computationally expensive for large datasets.


```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
# Generate sample data
X, y = make_blobs(n_samples=30, centers=3, n_features=2, random_state=42)
# Create DBSCAN object
dbscan = DBSCAN(eps=1.5, min_samples=4)
# Fit model and predict cluster labels
labels = dbscan.fit_predict(X)
# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.show()
print("Number of outliers:", sum(labels==-1))
```
